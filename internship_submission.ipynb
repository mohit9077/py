```python
post_type_strength = {
   'artwork': 4.0,
   'blog': 3.0, 
   'skill':2.0, 
   'project': 1.0,
     
}
```

```python
combined['post_Strength'] = combined[' post_type'].apply(lambda x: post_type_strength[x])
```

```python
combined_df = combined.drop_duplicates()
combined_df = combined_df.groupby(['user_id', 'post_id', 'title']).sum().reset_index()
```

```python
combined_df.head(n=50)
```

```python
combined_df.dtypes
```

```python
combined_df['title'] = combined_df['title'].astype("category")
combined_df['user_id'] = combined_df['user_id'].astype("category")
combined_df['post_id'] = combined_df['post_id'].astype("category")
combined_df['userid'] = combined_df['user_id'].cat.codes
combined_df['postid'] = combined_df['post_id'].cat.codes
```

```python
combined_df.head()
```

```python
sparse_post_user = sparse.csr_matrix((combined_df['post_Strength'].astype(float), (combined_df['postid'], combined_df['userid'])))
sparse_user_post = sparse.csr_matrix((combined_df['post_Strength'].astype(float), (combined_df['userid'], combined_df['postid'])))

model = implicit.als.AlternatingLeastSquares(factors=20, regularization=0.1, iterations=50)

alpha = 15
data = (sparse_post_user * alpha).astype('double')
model.fit(data)
```

```python
combined_df['postid']
```

```python
post_id = 20
n_similar = 10

user_vecs = model.user_factors
post_vecs = model.item_factors

post_norms = np.sqrt((post_vecs * post_vecs).sum(axis=1))

scores = post_vecs.dot(post_vecs[post_id]) / post_norms
top_idx = np.argpartition(scores, -n_similar)[-n_similar:]
similar = sorted(zip(top_idx, scores[top_idx] / post_norms[post_id]), key=lambda x: -x[1])

for post in similar:
    idx, score = post
    print(combined_df.title.loc[combined_df.postid == idx].iloc[0])
```

```python
def recommend(user_id, sparse_user_post, user_vecs, post_vecs, num_posts=10):
    # Get the interactions scores from the sparse user post matrix
    user_interactions = sparse_user_post[user_id,:].toarray()
    # Add 1 to everything, so that articles with no interaction yet become equal to 1
    user_interactions = user_interactions.reshape(-1) + 1
    # Make articles already interacted zero
    user_interactions[user_interactions > 1] = 0
    # Get dot product of user vector and all post vectors
    rec_vector = user_vecs[user_id,:].dot(post_vecs.T).toarray()
    
    # Scale this recommendation vector between 0 and 1
    min_max = MinMaxScaler()
    rec_vector_scaled = min_max.fit_transform(rec_vector.reshape(-1,1))[:,0]
    # post already interacted have their recommendation multiplied by zero
    recommend_vector = user_interactions * rec_vector_scaled
    # Sort the indices of the post into order of best recommendations
    post_idx = np.argsort(recommend_vector)[::-1][:num_posts]
    
    # Start empty list to store titles and scores
    titles = []
    scores = []

    for idx in post_idx:
        # Append titles and scores to the list
        titles.append(combined_df.title.loc[combined_df.postid == idx].iloc[0])
        scores.append(recommend_vector[idx])

    recommendations = pd.DataFrame({'title': titles, 'score': scores})

    return recommendations
    
# Get the trained user and post vectors. We convert them to csr matrices
user_vecs = sparse.csr_matrix(model.user_factors)
post_vecs = sparse.csr_matrix(model.item_factors)

# Create recommendations for user with id 50
user_id = 50

recommendations = recommend(user_id, sparse_user_post, user_vecs, post_vecs)

print(recommendations)
```

```python
def get_similar_posts(title,post_strength):
    
    similar_score=similarity_df[title]*(post_strength-2.5)
    similar_score=similar_score.sort_values(ascending=False)
    
    return similar_score
```
